```python
import numpy as np
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
```

```python
X[30:40]
```

```python
len(X)
```

```python
# defininf sine and cosine function
C, S = np.cos(X), np.sin(X)
```

```python
type(C)
```

```python
C[10]
```

```python
C[:10]
```

```python
# time to plot
import matplotlib.pyplot as plt
```

```python
plt.plot(X,C)
```

```python
plt.plot(X,S)
```

```python
plt.show() # hummm
```

```python
# creating a new figure of size 8x6, using 100 dots per inch
plt.figure(figsize=(8,6), dpi=80)
plt.plot(X,S)
```

```python
# creating a new subplot from a grid of 1x1
plt.subplot(111)
plt.plot(X,C)
```

```python
# now ploting with colours
plt.plot(X,C, color='blue', linewidth=1.0, linestyle='-')
```

```python
# setting limits
plt.xlim(-4.0, 4.0)
plt.plot(X,C, color='green', linewidth=2.0, linestyle='-')
```

```python
# setting ticks
plt.yticks(np.linspace(-1,1,20, endpoint=True))
plt.plot(X,S)
```

```python
# using spines
ax = plt.gca()
ax.spines['right'].set_color('red')
plt.plot(X,C, color='green', linewidth=3.0, linestyle='-')
```

```python
# adding legend and label
plt.plot(X,S, color='blue', linewidth=2.5, linestyle='-', label='sine')
plt.legend(loc='upper left', frameon=None)
```

#### regular plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2*X)
```

```python
plt.plot(X, Y+1, color='blue', alpha=1.0)
plt.plot(X, Y-1, color='blue', alpha=1.0)
```

```python
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
```

```python
plt.scatter(X,100*Y) # check the scales
```

```python
n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5, 1.0, n)
```

```python
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='red')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='green')
```
